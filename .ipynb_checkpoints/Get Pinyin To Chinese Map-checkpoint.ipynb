```python
def filter_cols(cols):
    res = []
    for col in cols:
        if col == "":
            continue
        if 'Unnamed' in col:
            continue
        # For following cases, check if they exists in `pinyin_chinese_map`
        # if so, replace them with chinese characters
        # if not, ignore them
        col_lower = col.lower()
        if re.sub(r'[a-z0-9_]', '', col_lower) == "":
            res.append(col_lower)
        else:
            res.append(col[:50])
    return res
```

```python
def get_xls_cols(filename):
    xl_workbook = xlrd.open_workbook(filename, on_demand = True)
    file_cols = []
    for sheet_name in xl_workbook.sheet_names():
        xl_sheet = xl_workbook.sheet_by_name(sheet_name)
        if xl_sheet.nrows > 0:
            # Get header & content cell_values
            cols = np.array(xl_sheet._cell_values[0]).astype(str)
            cols = filter_cols(cols)
            file_cols.extend(cols)
            # 某些excel表格中，表头会出现在前三行，进行以下判断保证所有列名被涵盖在内
            if len(cols) == 1 and xl_sheet.nrows > 1:
                cols = np.array(xl_sheet._cell_values[1]).astype(str)
                cols = filter_cols(cols)
                file_cols.extend(cols)
            if len(cols) == 1 and xl_sheet.nrows > 2:
                cols = np.array(xl_sheet._cell_values[2]).astype(str)
                cols = filter_cols(cols)
                file_cols.extend(cols)
    xl_workbook.release_resources()
    return file_cols
```

```python
def get_csv_cols(filename):
    if filename in {'../input/train/社区服务中心.csv', '../input/test2/3a005470f08021e03ccd9317d23021b9.csv'}:
        df = pd.read_csv(filename, nrows=100, usecols = range(4), error_bad_lines=False).reset_index().fillna("")
    else:
        df = pd.read_csv(filename, nrows=100, error_bad_lines=False).reset_index().fillna("")
    cols = list(df)
    cols = filter_cols(cols)
    return cols
```

```python
def preprocess_get_cols(df):
    """数据预处理,读取所有xls和csv文件,返回所有出现的columns"""
    all_cols = {}
    for _, row in tqdm_notebook(df.iterrows(), total=df.shape[0]):
    # for _, row in tqdm_notebook(df.head(2000).tail(10).iterrows(), total=10):
        filename = rf"../input/{row['filename']}"
        if filename[-3:] == 'xls':
            try:
                cols = get_xls_cols(filename)
            except:
                try:
                    cols = get_csv_cols(filename)
                except (pd.io.common.EmptyDataError, XLRDError, UnicodeDecodeError, pd.errors.ParserError, CompDocError):
                    print(filename)
        elif filename[-3:] == 'csv':
            try:
                cols = get_csv_cols(filename)
            except (pd.io.common.EmptyDataError, XLRDError, UnicodeDecodeError, pd.errors.ParserError, CompDocError):
                print(filename)
        all_cols[row['filename']] = cols
    return all_cols
```

```python
# 建立inverted dictionary，key是lowercase的列明，values是文件名的list，方便查找出现过列名的文件名
inv_dict = defaultdict(list)
for filename, cols in train_cols.items():
    for col in cols:
        col = reserve_chinese_title(col).lower()
        inv_dict[col].append(filename)
for filename, cols in test1_cols.items():
    for col in cols:
        col = reserve_chinese_title(col).lower()
        inv_dict[col].append(filename)
for filename, cols in test2_cols.items():
    for col in cols:
        col = reserve_chinese_title(col).lower()
        inv_dict[col].append(filename)
```

```python
%%time
# 统计列名出现的次数
c = Counter()
for filename, cols in train_cols.items():
    for col in cols:
        col = reserve_chinese_title(col).lower()
        c[col] += 1
for filename, cols in test1_cols.items():
    for col in cols:
        col = reserve_chinese_title(col).lower()
        c[col] += 1
for filename, cols in test2_cols.items():
    for col in cols:
        col = reserve_chinese_title(col).lower()
        c[col] += 1
```

```python
%%time
pinyin_dict = defaultdict(list)
# c是column出现的次数
for col, _count in c.items():
    if col == 'code0002' or col == 'error无访问权限' or col == 'index':
        continue
    if col and len(re.sub(r'[a-z0-9_]', '', col)) > 1:
        try:
            pinyin_result = get_pinyin_first_alpha(col)
            if pinyin_result == col:
                continue
            # 只有在拼音出现过的时候才保存
            if pinyin_result in c:
                pinyin_dict[pinyin_result].append((col, _count, c[pinyin_result]))
        except:
            continue
```

```python
len(pinyin_dict)
```

```python
unique_dict = {}
multiple_dict = {}
for k,v in pinyin_dict.items():
    if len(v) == 1:
        unique_dict[k] = v
    else:
        multiple_dict[k] = v
```

```python
# 保留规则:
# 拼音出现次数超过2次
# 拼音出现次数为2次且 len(pinyin) > 2
# 拼音出现次数为1次且 len(pinyin) > 3

pinyin_chinese_map = {}
for k, v in unique_dict.items():
    pinyin_count = v[0][2]
    if pinyin_count > 2 or (pinyin_count == 2 and len(k) > 2) or (pinyin_count == 1 and len(k) > 3):
        pinyin_chinese_map[k] = v[0][0]
```

```python
## 如果第一匹配出现的次数超过第二匹配的十倍，则认定第一匹配为真是匹配
picked_dict = {}
for k,v in multiple_dict.items():
    if 1.0 * v[0][1] / v[1][1] >= 10:
        picked_dict[k] = v[0][0]
```

```python
pinyin_chinese_map.update(picked_dict)
```

```python
pop_items = []
for k,v in pinyin_chinese_map.items():
    if k[0] == 'b' and k[1] != 'z' and v[0] == '表':
        pop_items.append(k)
```

```python
for item in pop_items:
    pinyin_chinese_map.pop(item)
```

```python
add_tokens = []
for k,v in c.items():
    # 规则：
    # 只包含英文字符和下划线
    # 出现超过10次
    # 不在pinyin_chinese_map中
    # 拼音长度超过两个字符
    if re.sub(r'[a-z_]', '', k) == "" and v > 10 and k not in pinyin_chinese_map and len(k) > 2:
        add_tokens.append(k)
```

```python
import pickle
with open("../tmp_input/pinyin_chinese_map_1201.pkl", 'wb') as f:
    pickle.dump(pinyin_chinese_map, f)
with open("../tmp_input/add_tokens_1201.pkl", 'wb') as f:
    pickle.dump(add_tokens, f)
```
